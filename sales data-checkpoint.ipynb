```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import textwrap

url="https://themeforest.net/top-sellers"
response=requests.get(url)
soup=BeautifulSoup(response.text, 'html.parser')
theme=soup.find_all('div', class_='shared-item_cards-grid-image_card_component__content')
data=[]
for i in theme:
    title=i.find('h3', class_="shared-item_cards-item_name_component__root").text.strip()
    author=i.find('a', class_='shared-item_cards-author_category_component__link').text.strip()
    price= i.find('div', class_='shared-item_cards-price_component__root').text.strip()
    sales_count=i.find('div', class_='shared-item_cards-sales_component__root').text.strip()
    data.append({
        'Theme Name':title,
        'Author':author,
        'Price':price,
        'Sales':sales_count
    })
print(data)
```

```python
# Convert data into a pandas DataFrame
df = pd.DataFrame(data)

#cleaning Data
df['Price'] = df['Price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['Sales'] = df['Sales'].replace({' Sales': '', 'K': '000'}, regex=True).astype(int)

#find null values
df.isnull().sum()
```

```python
#Price vs Sales Analysis (Scatter plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Sales', data=df)
plt.title('Price vs Sales')
plt.xlabel('Price ($)')
plt.ylabel('Sales')
plt.show()
```

```python
top_5_sales = df.nlargest(5, 'Sales')
plt.figure(figsize=(10, 6))
sns.barplot(x='Sales', y='Theme Name', data=top_5_sales, palette='viridis')
plt.title('Top 5 Most Sold Themes')
plt.xlabel('Sales')
plt.ylabel('Theme Name')
plt.show()
```

```python
#Price Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True, color='blue', bins=10)
plt.title('Price Distribution of Themes')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Price-to-Sales Ratio (Bar chart)
df['Price-to-Sales Ratio'] = df['Price'] / df['Sales']
plt.figure(figsize=(10, 6))
sns.barplot(x='Price-to-Sales Ratio', y='Theme Name', data=df, palette='magma')
plt.title('Price-to-Sales Ratio of Themes')
plt.xlabel('Price-to-Sales Ratio')
plt.ylabel('Theme Name')
plt.show()
```

```python
#Sales Breakdown by Theme (Bar chart)
plt.figure(figsize=(10, 6))
sns.barplot(x='Sales', y='Theme Name', data=df, palette='Blues')
plt.title('Sales Breakdown by Theme')
plt.xlabel('Sales')
plt.ylabel('Theme Name')
plt.show()
```
